# Testing Ground

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Get all available color palette names
palette_names = list(sns.palettes.SEABORN_PALETTES.keys())

# Plot each palette
num_palettes = len(palette_names)
fig, axes = plt.subplots(nrows=num_palettes, figsize=(8, num_palettes * 0.5))

for ax, palette_name in zip(axes, palette_names):
    # Get the colors for each palette
    colors = sns.color_palette(palette_name)
    
    # Plot the palette manually on each subplot
    ax.imshow([colors], aspect='auto')
    ax.set_title(palette_name, fontsize=12)
    ax.set_xticks([])
    ax.set_yticks([])

plt.tight_layout()
plt.show()

```
